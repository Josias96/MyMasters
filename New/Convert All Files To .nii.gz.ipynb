```python
#------------------------------------
# Just Run This Cell -> Shift + Enter
#------------------------------------

import os
from os.path import isfile, join

import numpy as np
import SimpleITK as sitk
```

```python
#------------------------------------
# Just Run This Cell -> Shift + Enter
#------------------------------------

def CompressFiles(DataPath, RemoveOldFiles=False):
    """Set RemoveOldFiles to True if old files are to be removed. """ 
    testString1 = ".nii.gz"
    testString2 = ".nii"
    testString3 = ".hdr"
    testString4 = ".img"
    try:
        for subdir, dirs, files in os.walk(DataPath):
            for file in files:
                if testString1 in file:
                    pass
                elif testString2 in file:
                    try:
                        filePath = join(subdir,file).replace("\\","/")
                        SITK_IMG  = sitk.ReadImage(filePath)
                        newFileName = file.replace(testString2,testString1)
                        sitk.WriteImage(SITK_IMG,join(subdir,newFileName))
                        print("Sucess: .nii to .nii.gz")
                        try:
                            if RemoveOldFiles:
                                os.remove(filePath)
                            else:
                                pass
                        except:
                            print("Error Removing Uncompressed File")
                    except:
                        print("Error Writting: .nii to .nii.gz \t File: " + join(subdir,file))
                elif testString3 in file:
                    try:
                        filePath = join(subdir,file).replace("\\","/")
                        additional_filepath = filePath.replace(testString3, testString4)
                        SITK_IMG  = sitk.ReadImage(filePath)
                        newFileName = file.replace(testString3,testString1)
                        sitk.WriteImage(SITK_IMG,join(subdir,newFileName))
                        print("Sucess: .img to .nii.gz")
                        try:
                            if RemoveOldFiles:
                                os.remove(filePath)
                                os.remove(additional_filepath)
                            else:
                                pass
                        except:
                            print("Error Removing Uncompressed File")
                    except:
                        print("Error Writting: .img to .nii.gz \t File: " + join(subdir,file)) 
                else:
                    pass
    except:
        print("Error: Bad Input Directory!")
```

```python
#------------------------------------
#Change dir_path to the path containig all folders with the desired scans to convert
#------------------------------------
dir_path = "F://MyMasters//Data//Root"
CompressFiles(dir_path, RemoveOldFiles=True)
```
